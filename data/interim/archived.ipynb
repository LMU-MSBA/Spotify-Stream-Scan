```python
def fetch_artists(genre, limit=50, total=1000):
    artists = set()  # Using a set to avoid duplicate artists
    offset = 0  # Start at the beginning of the results
    request_count = 0  # Keep track of the number of requests made

    while len(artists) < total:
        print(f"Fetching batch {request_count + 1} (offset {offset})...")  # Log the current batch and offset
        # Perform the search
        result = spotify.search(q=f'genre:"{genre}"', type='artist', limit=limit, offset=offset)
        batch_artists = [artist['name'] for artist in result['artists']['items']]
        artists.update(batch_artists)  # Add to the set of artists

        print(f"Fetched {len(batch_artists)} artists. Total unique artists collected: {len(artists)}")

        if not batch_artists:
            print("No more artists found, ending search.")
            break  # Exit if no more artists are returned

        offset += len(batch_artists)  # Prepare the offset for the next batch of results
        request_count += 1  # Increment the request count

        if len(artists) >= total:
            print(f"Reached the target of {total} artists, ending search.")
            break  # Break the loop if we have collected enough artists

    return list(artists)[:total]  # Return up to the requested number of artists

# Fetch 1000 pop artists
pop_artists = fetch_artists('pop', total=1000)
print(f"Number of artists fetched: {len(pop_artists)}")
# Optionally, print the list of artists
# for artist in pop_artists:
#     print(artist)
```

```python
import time
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Assuming Spotify client setup
# spotify = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials(client_id='YOUR_CLIENT_ID', client_secret='YOUR_CLIENT_SECRET'))

def get_album_tracks(album_ids):
    all_tracks_data = []
    processed_albums = set()  # Track processed album IDs to avoid duplicates

    def process_batch(batch):
        for album_id in batch:
            if album_id in processed_albums:
                print(f"Album ID {album_id} has already been processed. Skipping.")
                continue
            print(f"Processing album ID: {album_id}")
            try:
                tracks = spotify.album_tracks(album_id)
                track_items = tracks['items']
                print(f"Found {len(track_items)} tracks in album ID {album_id}.")
                
                for track in track_items:
                    all_tracks_data.append({
                        'Album ID': album_id,
                        'Track Name': track['name'],
                        'Track ID': track['id'],
                        'Artists': ', '.join(artist['name'] for artist in track['artists']),
                        'Duration (ms)': track['duration_ms'],
                        'Track Number': track['track_number']
                    })
                processed_albums.add(album_id)  # Mark as processed
            except spotipy.SpotifyException as e:
                print(f"An error occurred while processing album ID {album_id}: {e}")

    # Process in batches
    batch_size = 20  # Define your batch size
    for i in range(0, len(album_ids), batch_size):
        batch = album_ids[i:i + batch_size]
        process_batch(batch)

    return pd.DataFrame(all_tracks_data)

# Replace 'album_ids' with your list of album IDs
# Example usage
# album_ids = ['album_id1', 'album_id2', ...]
# tracks_df = get_album_tracks(album_ids)
# print(tracks_df)
```
