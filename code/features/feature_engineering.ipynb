```python
genres = pd.read_csv('genres.csv')
artist_info = pd.read_csv('pop_artists_info.csv')
artist_info = artist_info[['Artist_ID', 'Popularity']]

df = pd.merge(genres, artist_info, on = 'Artist_ID', how = 'left')
df.head()
```

```python
grouped_df = df.groupby(['Artist_ID', 'Artist_Name', 'Popularity'])['Genre'].agg(list).reset_index()
grouped_df

one_hot_encoded_df = pd.get_dummies(grouped_df.explode('Genre'), columns = ['Genre'], prefix = '', prefix_sep = '')
one_hot_encoded_df.head()
one_hot_encoded_df = one_hot_encoded_df.replace({True: 1, False: 0})
one_hot_encoded_df = one_hot_encoded_df.groupby(['Artist_ID', 'Artist_Name', 'Popularity']).sum()
one_hot_encoded_df.head()
```
