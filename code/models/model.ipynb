# Spotify Stream Scan - Reccomender Systems Model Development

```python
# get the required modules

import pandas as pd                     # get the pandas
import numpy as np
from sklearn.neighbors import NearestNeighbors
import pickle
```

```python
df = pd.read_csv('model_df.csv')
df = df.iloc[:,1:]
df.head()
```

```python
# Selecting the relevant features (Rating + Genre Encodings)
model_df = df.iloc[:,2:550]

# Fitting the kNN model
neigh = NearestNeighbors(n_neighbors=5)
neigh.fit(model_df)
```

```python
prediction_df = pd.DataFrame(columns = model_df.columns)
prediction_df.loc[0] = [0] * len(prediction_df.columns)

prediction_df['pop'] = 1
prediction_df['r&b'] = 1
prediction_df['rap'] = 1

prediction_df
```

```python
neighbors = neigh.kneighbors(prediction_df)
neighbors
```

```python
# Indices of 5 nearest neighbors in df
arr = np.array(neighbors[1])
recs = df.iloc[arr[0], 0:2].sort_values(by = 'Popularity', ascending = False)

print(recs)
```

```python
with open('nearest_neighbors_outputs.pkl', 'wb') as f:
    pickle.dump(neighbors, f)
```
